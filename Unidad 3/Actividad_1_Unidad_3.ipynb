```python
import numpy as np
from scipy.stats import binom


# Parametros
n = 3 # numero de volados (ensayos)
p = 0.5 # probabilidad de tener sol en cada volado (éxito)

# Valores de X
x_values = [0, 1, 2, 3]

# Calcular la PMF(Funcion de Masa de Probabilidad) para cada valor de X
pmf_values = binom.pmf(x_values, n, p)

# Imprimir los resultados
for x, pmf in zip(x_values, pmf_values):
    print(f'P(X={x}) = {pmf:.3f}')
```

# **Actividad 2**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parametros
n = 3 # numero de volados (ensayos)
p = 0.5 # probabilidad de tener sol en cada volado (éxito)

# Distribucion binomi
x = np.arange(0, n+1)
pmf = binom.pmf(x, n, p)

# Grafica
plt.bar(x, pmf)
plt.xlabel('Numero de caras')
plt.ylabel('Probabilidad')
plt.title('Distribucion Binomial')
plt.show()
```
