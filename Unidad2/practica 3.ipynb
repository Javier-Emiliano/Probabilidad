```python
# Paso 1: importar librerias
import math as mt
from itertools import permutations
from itertools import product
import numpy as np
```

```python
mt.factorial(10)
```

```python
mt.factorial(10) / mt.factorial(10-4)
```

```python
# Paso 2: Comenzamos con la premunición
nombres = ("Javier", "Rocio", "Saúl", "Ruby", "Lucy", "José", "Luna", "Mateo", "Lucas", "Lily")
n = len(nombres)
r = 4 # Agrupación

Pr = mt.factorial(n)/mt.factorial(n-r)
print("Existen", Pr , "posibles premuniciones sin repetición total de", n , "nombres", "en grupos de", r , "en", r)

print("aqui premuniciones")
premuniciones_nombres = list(permutations(nombres,r))

# Paso 3: imprime las premuniciones
for permutacion in premuniciones_nombres:
  print(permutacion)
```
