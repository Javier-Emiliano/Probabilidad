```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# Paso 1: Generar datospara la distribución normal


media = 70 # Media
desv_std = 4 # Desvición estándar
num_datos = 500

# Paso 2: Generar datos de manera aleatoria y los ponemos en una lista (array)
datos = np.random.normal(media, desv_std, num_datos)
datos[0:50]
```

```python
# Paso 3: Graficar histograma
plt.figure(figsize=(6, 4))
plt.hist(datos, bins=15, alpha=0.4, color='b') # Histograma de los datos
plt.title('Distribución Normal')
plt.xlabel('Peso en kg de los estudiantes')
plt.grid(True)
plt.show()
```

```python
# Paso 4: Crear un rango de valores para crear una densidad
x = np.linspace(media - 3 * desv_std, media + 3 * desv_std, num_datos)

# Paso 5: Calcular el PDF de la distribución normal
pdf = 1 / (desv_std * np.sqrt(2 * np.pi)) * np.exp(-(x - media)**2 / (2 * desv_std**2))
```

```python
pdf.size, num_datos
```

```python
# Paso 6: Visualizar la PDF
plt.figure(figsize=(6, 4))
plt.plot(x, pdf, label='Normal:(media:{}; ds:{})'.format(np.round(media, 2), np.round(desv_std, 2)))
plt.title('Funcion de Densidad de Probabilidad (PDF) - Distribución Normal')
plt.xlabel('Pesos en kg. de los estudiantes')
plt.ylabel('Densidad de probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Paso 7: Graficar Histograma
plt.figure(figsize=(6, 4))
plt.hist(datos, bins=15, alpha=0.4, density=True, color='b')
plt.plot( x, pdf, 'k', linewidth=2, )

plt.title('Distribución Normal')
plt.xlabel('Peso en kg de los estudiantes')
plt.grid(True)
plt.show()
```

# **Ejercicio 2**

Una revisión de los pesos de **500 estudiantes del ITD**, evidencio que la **Media es de 70 kilogramos** y la **Desviación estadar es de 4 kilogramos**.Suponiendo que los valores se distribuyen normlmente. Hallar lo siguiente:

*   Cuantos estudiantes pesan menos de 75 kg.
*   Cuantos estudiantes pesan menos de 65 kg.
*   Cuantos estudiantes pesan más de 79 kg.

Primero necesitamos calcular la probabilidad de que un estudiante pese menos de 75 kg utilizando la funcion de distribucion acumulativa (CDF) de la distribución normal






```python
plt.figure(figsize=(6, 4))

# Parametros de la distribución normal
media = 70 #Media
desv_std = 4 # Desviasión estándar
total_estudiantes = 500
#VARIABLE para el Cálculo de la probabilidad de que un estudiante pese menos de 75 kg
peso_limite = 79
# Paso 1: Calcular la probablilidad de que un estudiante pese menos de 75 kg usando la funcion de distribución acumulativa (CDF) de la distribución normal
prob_menos_75kg = norm.cdf(peso_limite, media, desv_std)

# Paso 2: Calcular la probabilidad de que un estudiante pese menos de 65 kg
estudiantes_menos_75kg = total_estudiantes * prob_menos_75kg

print("Probabilidad de que un estudiante pese menos de 75 kg:", prob_menos_75kg)
print("Número de estudiantes que pesan menos de 75 kg:", estudiantes_menos_75kg)
# Paso 3: Graficar la distribución normal
x = np.linspace(media - 4 * desv_std, media + 4 * desv_std, 1000)
y = norm.pdf(x, media, desv_std)

plt.plot(x, y, 'b', linewidth=2)
plt.fill_between(x, y, where=(x <= peso_limite), color='skyblue', alpha=0.5)
plt.title('Probabilidad de que los estudiantes pesen menos de 75 kg.\n $\mu=70$, $\sigma=4$, X = 500')
plt.xlabel('Peso en (kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```

**¿Cuantos estudiantes pesan menos de 62 kg?**

```python
plt.figure(figsize=(6, 4))

# Parámetros e la distribución normal
media = 70 # Media
desv_std = 4 # Desviación estándar
total_estudiantes = 500
# VARIABLE para el cálculo de la probabilidad de que un estudiante pese menos de 75 kg
peso_limite = 62
# Paso 1: Calcular la probabilidad de que un estudiante pese menos de 75 kg utilizando la función de distribución acumulativa (CDF) de la distribución normal
prob_menos_62kg = norm.cdf(peso_limite, media, desv_std)

# Paso 2: Calcular la probabilidad de que un estudiante pese menos de 65 kg
estudiantes_menos_62kg = total_estudiantes * prob_menos_62kg

print("Probabilidad de que un estudiante pese menos de 62 kg:", prob_menos_62kg)
print("Número de estudiantes que pesan menos de 62 kg:", estudiantes_menos_62kg)

# Paso 3: Calcular la distribución normal
x = np.linspace(media - 4 * desv_std, media + 4 * desv_std, total_estudiantes)
y = norm.pdf(x, media, desv_std)

# Paso 4: Graficar la distribución normal
plt.plot(x, y, 'b', linewidth=2)
plt.fill_between(x, y, where=(x <= peso_limite), color='skyblue', alpha=0.5)
plt.title('Probabilidad de que los estudiantes pesen menos de 62 kg.\n $\mu=70$, $\sigma=4$, X = 500')
plt.xlabel('Peso en (kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```

**¿Cuántos estudiantes pesan más de 79 kg?**

```python
plt.figure(figsize=(6, 4))

# Parámetros e la distribución normal
media = 70 # Media
desv_std = 4 # Desviación estándar
peso_limite = 79 # Peso límite
total_estudiantes = 500

# Paso 1: Calcular la probabilidad de que un estudiante pese menos o igual a 79 kg
prob_menos_79kg = norm.cdf(peso_limite, media, desv_std)

# La probabilidad de que un estudiante pese más de 79 kg es complementaria
prob_más_79kg = 1 - prob_menos_79kg

# Paso 2: Calculamos el número aproximado de estudiantes que pesan más de 79 kg
estudiantes_más_79kg = total_estudiantes * prob_más_79kg

print("Probabilidad de que un estudiante pese más de 79 kg:", prob_más_79kg)
print("Número aproximado de estudiantes que pesan más de 79 kg:", estudiantes_más_79kg)

# Paso 3: Calcular la distribución normal
x = np.linspace(media - 4 * desv_std, media + 4 * desv_std, total_estudiantes)
y = norm.pdf(x, media, desv_std)

# Paso 4: Graficar la distribución normal
plt.plot(x, y, 'b', linewidth=2)
plt.fill_between(x, y, where=(x >= peso_limite), color='skyblue', alpha=0.5)
plt.title('Probabilidad de que los estudiantes pesen más de 79 kg.\n $\mu=70$, $\sigma=4$, X = 500')
plt.xlabel('Peso en (kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```

# **Distribución Binomial**
La probabilidad de éxito de una vacuna es de 0.83. Calcular la probabilidad de que una vez administrada a 16
pacientes NINGUNO sufra la enfermedad.

```python
from scipy.stats import binom

# Parámetros de la distribución binomial
probabilidad_éxito = 0.83
num_pacientes = 16
X = 16 # Número de éxitos

# Paso 1: Calcular la probabilidad de que ningún paciente sufra la enfermedad
probabilidad_ninguno_enfermo = binom.pmf(X, num_pacientes, probabilidad_éxito)

print("Probabilidad de que ningún paciente sufra la enfermedad:", probabilidad_ninguno_enfermo)
```

La probabilidad de éxito de una vacuna es de 0.83. Calcular la probabilidad de que una vez administrada a 16
pacientes TODOS sufra la enfermedad.

```python
# Parámetros de la distribución binomial
probabilidad_éxito = 0.83
num_pacientes = 16
X = 0 # Número de éxitos

# Paso 1: Calcular la probabilidad de que ningún paciente sufra la enfermedad
probabilidad_todos_enfermo = binom.pmf(X, num_pacientes, probabilidad_éxito)

print("Probabilidad de que todos paciente sufra la enfermedad:", probabilidad_todos_enfermo)
```

# **Ejercicio 2**


*   Para una variable aleatoria binomial con n= 10 y una probabilidad de 0.10
*   Encuentra la probabilidad de la variable aleatoria x= 2



```python
from scipy.stats import binom

# Parámetros de la distribución binomial
n = 10
p = 0.10
x = 2 # Número de éxitos

# Paso 1: Calcular la probabilidad de la variable aleatoria x= 2
probabilidad_x_2 = binom.pmf(x, n, p)

print("Probabilidad de la variable aleatoria x sea igual a 2 es:", probabilidad_x_2)
```

# **Ejercicio 3**
*   Para una variable aleatoria binomial con n= 10 y una probabilidad de 0.10
*   Calcular las probabilidades para cada valor de x = 0 hasta 10
*   Imprimir la tabla de valores de x y sus probabilidades
*   Graficar el histograma de la distribución binomial



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parámetros de la función binomial
n = 10
p = 0.10

# Calcular las probabilidades para cada valor de x de 0 a 10
valores_x = np.arange(0, n+1)
probabilidades_x = [binom.pmf(x, n, p) for x in valores_x]

# Imprimir la tabla de valores de x y sus probabilidades
print("Valores de x\tProbabilidades")
for x, probabilidad in zip(valores_x, probabilidades_x):
    print(f"{x}\t\t{probabilidad}")
```

```python
# Graficar el histograma de la distibución normal
plt.figure(figsize=(8, 6))
plt.bar(valores_x, probabilidades_x, color='skyblue')
plt.title('Distribución Binomial')
plt.xlabel('Valor de X')
plt.ylabel('Probabilidad')
plt.xticks(valores_x)
plt.grid(True)
plt.show()
```

# **Distribución Poisson**


*   Una compañía de seguros de vida asegura la vida de **5000 hombres** de **42
    años** de edad
*   Si los estudios actuariales muestran que **la probabilidad de que cualquier
    hombre de 42 años muera en un año determinado es de .001**. Encuentra la probabilidad exacta de que la compañía **tendrá que pagar x = 4 durante un año determinado**



```python
from scipy.stats import poisson

# Paso 1: Parámetros de la distribución de poisson
media_muertes = 5000 * 0.001 # 5000 hombres con una probabilidad de muerte de 0.001

# Paso 2: Sacar el número de reclamaciones
x = 4

# Paso 3: Calcular la probailidad de que la compañía tenga que pagar 4 reclamaciones en un año determinado
probabilidad_4_reclamaciones = poisson.pmf(x, media_muertes)

print("Probabilidad de que la compañía pague 4 reclamaciones es DE:",
      probabilidad_4_reclamaciones)
```
